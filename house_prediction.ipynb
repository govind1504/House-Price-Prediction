# IMPROTING LIBRARIES AND DATASET

```python
#importing Libraries
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/boston.csv')
df.head()
```

# DATA REVIEW

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

# MODEL USING SINGLE FEATURE (LSTAT)

```python
#Single feature comp with medv (Feature will be lstat)
data = df[['LSTAT','MEDV']]
data
```

VISUALIZTION OF MEDV VS LSTAT

```python
data.plot(x='LSTAT',y='MEDV',style='o')
plt.xlabel('istat')
plt.ylabel('medv')
plt.show()
```

# MODEL CREATION

```python
#different the dependent and independent features
x = pd.DataFrame(data['LSTAT'])
y = pd.DataFrame(data['MEDV'])
```

```python
#Divide data in train and test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=1 )
```

```python
print(f"X train shape is : {x_train.shape}")
print(f"Y train shape is : {y_train.shape}")
print("-"*30)
print(f"X test shape is :{x_test.shape}")
print(f"y test shape is :{y_test.shape}")
```

```python
#Model creation
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#fit the model
model.fit(x_train,y_train)
```

```python
#Coefficent
print(model.coef_)
```

```python
#interception
print(model.intercept_)
```

```python
#Predictions
y_prediction = model.predict(x_test)
y_prediction = pd.DataFrame(y_prediction , columns = ['Predicted'])
y_prediction.size
```

```python
y_prediction
```

# ERROR CALCULATION


1.   Mean Absolute Error
2.   Mean Squared Error
3.   Root Mean Squared Error
4.   R2 Score



```python
#Error Calculation
from sklearn import metrics

print(f'MAE = {metrics.mean_absolute_error(y_test,y_prediction)}')
print(f"MSE = {metrics.mean_squared_error(y_test,y_prediction)}")
print(f"RMSE = {np.sqrt(metrics.mean_squared_error(y_test,y_prediction))}")
```

```python
print(f"rSquared = {metrics.r2_score(y_test,y_prediction)}")
```

# Actual Vs Predict  Visualization

```python
#Actual Vs Predicted for single feature Model
plt.scatter(y_test, y_prediction)
plt.ylim(0,50)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```

# fine tuning using Multiple Linear Regression

```python
#Multiple Linear Regression
x = pd.DataFrame(df.iloc[:,:-1])
y = pd.DataFrame(df.iloc[:,-1])
```

Model Creation

```python
#divide data in train and test
X_train , X_test , Y_train , Y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#Create Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#fit the model
model.fit(X_train,Y_train)
```

```python
#coefficient
model.coef_
```

```python
#interception
model.intercept_
```

```python
#Predictions
y_prediction = model.predict(X_test)
y_prediction = pd.DataFrame(y_prediction , columns = ['Predicted'])
y_prediction.size
```

# Actual Vs Predicted visualization

```python
x = Y_test
y = y_prediction

plt.scatter(Y_test,y_prediction)
plt.ylim(0,50)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```
